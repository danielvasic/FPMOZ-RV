```python
from skimage import io
from skimage import img_as_float
```

```python
fpmoz = io.imread ("fpmoz.jpg", as_grey=True)
fpmoz = img_as_float (fpmoz)
```

```python
io.imshow(fpmoz)
```

```python
import numpy as np

sobel_x =np.array ([
    [1,0,-1],
    [2,0,-2],
    [1,0,-1]
]);

sobel_x
```

```python
from scipy import ndimage
fpmoz_out = ndimage.convolve (fpmoz, sobel_x, mode="mirror", cval=0.0)
fpmoz_out2 = ndimage.convolve (fpmoz, sobel_x.T, mode="mirror", cval=0.0)
```

```python
fpmoz_bw = fpmoz_out > 0.2
fpmoz_bw2 = fpmoz_out2 > 0.2
```

```python
io.imshow(fpmoz_bw)
```

```python
io.imshow(fpmoz_bw2)
```

```python
io.imshow(fpmoz_bw + fpmoz_bw2)
```
