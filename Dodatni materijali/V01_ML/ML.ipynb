```python
def loadDataset (filename, split=0.75):
    import csv, random
    trainSet=[]
    testSet=[]
    with open (filename, 'r') as csv_file:
        flowers = csv.reader(csv_file)
        dataset = list (flowers)
        for i in range (len(dataset)-1):
            for j in range(4):
                dataset[i][j] = float(dataset[i][j])
            if random.random() < split:
                trainSet.append(dataset[i])
            else:
                testSet.append(dataset[i])
        return trainSet, testSet
trainSet, testSet = loadDataset('iris.data.txt')
```

```python
def euclidianDistance (inst1, inst2, length):
    import math
    # sqrt(sum((x1 - y1)^2 + ... + (xn - yn)^2))
    distance = 0
    for i in range(length):
        distance += pow(inst1[i] - inst2[i], 2)
    return math.sqrt (distance)

a = [1,2,4,5,'a']
b = [2,2,6,7,'b']
euclidianDistance (a,b,4)
```

```python
def getNeighbours (trainSet, testInstance, k):
    import operator
    distances = []
    itemLen = len(testInstance)-1
    for i in range (len(trainSet)):
        dist = euclidianDistance (trainSet[i], testInstance, itemLen)
        distances.append ((trainSet[i], dist))
    distances.sort (key=operator.itemgetter(1))
    neighbours = []
    
    for i in range (k):
        neighbours.append (distances[i][0])
    return neighbours
a = [[1,2,4,5,'a'], [3,4,5,4,'b']]
b = [2,2,6,7]
getNeighbours(a,b,1)
```

```python
def getResponse (neighbours):
    import operator
    classVotes = {}
    for i in range(len(neighbours)):
        response = neighbours[i][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted (classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]

neighbors = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]
response = getResponse(neighbors)
print(response)
```

```python
def getAccuracy (testSet, predictions):
    correct = 0
    for i in range (len(testSet)):
        if testSet[i][-1] == predictions[i]:
            correct += 1
            
    return (correct / len(testSet)) * 100
```

```python
trainSet, testSet = loadDataset('iris.data.txt')
k = 3
predictions = []
for i in range (len(testSet)):
    neighbours = getNeighbours(trainSet, testSet[i], k)
    prediction = getResponse (neighbours)
    predictions.append (prediction)
getAccuracy (testSet, predictions)
```
