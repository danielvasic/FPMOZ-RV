```python
from skimage import io
from skimage import img_as_float
```

```python
fpmoz = io.imread('fpmoz.jpg', as_grey=True)
fpmoz = img_as_float(fpmoz)
```

```python
io.imshow(fpmoz)
```

```python
from skimage.util import random_noise
fpmoz_sum = random_noise(fpmoz, mode='speckle')
io.imshow(fpmoz_sum)
```
