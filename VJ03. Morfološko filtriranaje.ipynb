```python
from skimage import io
fpmoz = io.imread("fpmoz_zgrada.jpg", as_grey=True)
```

```python
maska = fpmoz > 0.5
```

```python
import matplotlib.pyplot as plt

def prikazi(orginal, obrada):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8,4), 
                                   sharex=True, sharey=True)
    ax1.set_title("Orginal")
    ax1.imshow(orginal, cmap=plt.cm.gray)
    ax1.axis("off")
    
    ax2.set_title("Obrada")
    ax2.imshow(obrada, cmap=plt.cm.gray)
    ax2.axis("off")
prikazi (fpmoz, maska)
```

```python
from skimage.filters import threshold_mean,  threshold_minimum
sredina =  threshold_mean(fpmoz)
sredina_maska = fpmoz > sredina
prikazi(fpmoz, sredina_maska)
```

```python
prikazi(maska, sredina_maska)
```

```python
from skimage.morphology import erosion, dilation
erosion_maska = erosion(sredina_maska)
dilation_maska = dilation(sredina_maska)
prikazi(sredina_maska, erosion_maska)
```

```python
prikazi(sredina_maska, dilation_maska)
```
